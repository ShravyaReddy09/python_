```python
#teeeeej....

class Menu:
    def __init__(self):
        self.items = {"1":{"item":"chicken biryani","price":240},"2":{"item":"chicken fry briyani","price":260},"3":{"item":"muttion briyani","price":300},
            "4":{"item":"fish briyani","price":350},"5":{"item":"water bottle","price":20},"6":{"item":"cool drinks","price":30},"7":{"item":"noodles","price":80}}

    def display_menu(self):
        print("Welcome to the pyspiders food court Menu:")
        for key, value in self.items.items():
            print(f"{key}. {value['item']}: {value['price']}")
            print()


class Cart:
    def __init__(self):
        self.items = []

    def add_to_cart(self, menu, item_number):
        if item_number in menu.items:
            self.items.append(menu.items[item_number])
            print(f"{menu.items[item_number]['item']} added to cart")
        else:
            print("Invalid item number. Please select a valid item Number.")

    def view_cart(self):
        total = sum(item['price'] for item in self.items)
        print("Your Cart:")
        for item in self.items:
            print(f"{item['item']}: {item['price']}")
        
        print(f"Total: {total}")

class Order:
    def __init__(self, cart):
        self.cart = cart

    def view_order_details(self):
        print("Order Details:")
        self.cart.view_cart()

    def process_payment(self):
        total = sum(item['price'] for item in self.cart.items)
        print(f"Payment processed amount of:{total}")


class User:
    def __init__(self, username, password):
        self.username = username
        self.password = password


class UserManager:
    def __init__(self):
        self.users = {}

    def register_user(self, username, password):
        if username not in self.users:
            self.users[username] = User(username, password)
            print("Registration successful!")
        else:
            print("Username already exists. Please choose a different username.")

    def login_user(self, username, password):
        if username in self.users and self.users[username].password == password:
            return self.users[username]
        else:
            print("Invalid username or password.")
            return None



user_manager = UserManager()

username = input("create a username: ")
password = input("create a password: ")
user_manager.register_user(username, password)  

username_input = input("Enter your username to login: ")
password_input = input("Enter your password to login: ")
logged_in_user = user_manager.login_user(username_input, password_input)  

if logged_in_user:
    print("Login successfull Welcome :", logged_in_user.username)
    print("your user_name:", logged_in_user.username)
    print("your user_password:", logged_in_user.password)
    food_menu = Menu()
    food_menu.display_menu()  
    cart = Cart()
    while True:
        choice = input("Enter the item number you want to add to the cart (or 'q' to quit): ")
        if choice.lower() == 'q':
            break
        cart.add_to_cart(food_menu, choice)
    cart.view_cart() 
    order = Order(cart)
    order.view_order_details()
    order.process_payment()
    total_cost = sum(item['price'] for item in cart.items)
    print("successfully your payment have done:",total_cost)
```
