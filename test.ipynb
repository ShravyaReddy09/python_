```python

    
class Employee:
    def __init__(self, name, position, salary):
        self.name = name
        self.position = position
        self.salary = salary

    def display_details(self):
        print(f"Name: {self.name}, Position: {self.position}, Salary: {self.salary}")

    @staticmethod
    def apply_salary_hike(current_salary):
        return current_salary * 1.04  # 4% salary hike

    def apply_hike_and_display(self):
        self.salary = self.apply_salary_hike(self.salary)
        self.display_details()


# Create instances of the Employee class
employee1 = Employee("John", "Software Engineer", 60000)
employee2 = Employee("Jane", "Data Scientist", 70000)

# Display initial details
print("Initial employee details:")
employee1.display_details()
employee2.display_details()

# Apply a 4% salary hike using the static method and display updated details
print("\nEmployee details after a 4% salary hike:")
employee1.apply_hike_and_display()
employee2.apply_hike_and_display()




          
```
