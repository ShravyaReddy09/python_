```python
class MenuItem:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name} - ${self.price}"

class Menu:
    def __init__(self):
        self.items = []

    def add_item(self, name, price):
        new_item = MenuItem(name, price)
        self.items.append(new_item)

    def display_menu(self):
        print("Menu:")
        for item in self.items:
            print(item)

class Order:
    def __init__(self):
        self.items = []

    def add_item(self, menu, item_name):
        for menu_item in menu.items:
            if menu_item.name == item_name:
                self.items.append(menu_item)
                print(f"{item_name} added to the order.")
                return
        print(f"{item_name} not found in the menu.")

    def display_order(self):
        if not self.items:
            print("Order is empty.")
            return
        print("Order:")
        for item in self.items:
            print(item)

class Restaurant:
    def __init__(self):
        self.menu = Menu()
        self.order = Order()

    def run(self):
        while True:
            print("\n1. Display Menu")
            print("2. Add Item to Order")
            print("3. Display Order")
            print("4. Exit")
            choice = input("Enter your choice (1-4): ")

            if choice == '1':
                self.menu.display_menu()
            elif choice == '2':
                item_name = input("Enter item name to add to order: ")
                self.order.add_item(self.menu, item_name)
            elif choice == '3':
                self.order.display_order()
            elif choice == '4':
                print("Thank you for using the Restaurant Management System. Exiting...")
                break
            else:
                print("Invalid choice. Please try again.")

# Example usage
restaurant = Restaurant()
restaurant.menu.add_item("Burger", 99)
restaurant.menu.add_item("Pizza", 150)
restaurant.menu.add_item("Pasta", 79)
restaurant.menu.add_item(" veg biryani", 199)
restaurant.menu.add_item("chicken biryani",249)
restaurant.menu.add_item("dum biryani", 249)
restaurant.menu.add_item("gongura chicken biryani", 299)
restaurant.run()



2 / 2





```

```python
class parent1:
    a=100
    b=50
    def display(self):
        print('parent1',self.a,self.b)

class parent2:
    a=25
    d=120
    def display(self):
        print('parent2',self.a,self.d)

class c(parent1,parent2):
    def display(self):
        super().display()
        parent2.display(self)

obj=c()
obj.display()
```
